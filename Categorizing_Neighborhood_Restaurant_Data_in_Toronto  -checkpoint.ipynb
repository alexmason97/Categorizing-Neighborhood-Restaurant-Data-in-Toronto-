```python
Url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(Url).text
```

```python
#utilize BeautifulSoup from bs4 to pull the data out 
#of the wiki page and bring it into our notebook
Bsoup = BeautifulSoup(source, "html.parser")
table = Bsoup.find('table')
```

If you noticed earlier, we had some values that weren't actually related to food such as "RBC" which was under the category "Bank". To fix this we will remove entries that are no associated with food/restaurant categories.

Next, we will One-Hot-Encode the Venue Categories. This in english basically meaning converting variables to numerical values in an interpretable format.

```python
TO_df_onehot = pd.get_dummies(TO_df[['Venue_Category']], prefix="", prefix_sep="")

TO_df_onehot['Neighborhood'] = TO_df['Neighborhood']

fixed_columns = [TO_df_onehot.columns[-1]] + list(TO_df_onehot.columns[:-1])
TO_df_onehot = TO_df_onehot[fixed_columns]

print("Shape of Dataframe:", TO_df_onehot.shape)
TO_df_onehot.head(10)
```

**Determine the total amount of venues of each category in each neighborhood**

```python

venue_count = TO_df_onehot.groupby('Neighborhood').sum()


venue_count.head(10)
```

### Data Visualization

Now we will determine the most popular restaurants by number of occurences and then use some bar graphs to display some the top neighborhoods with specified restaurant categories 

```python
top_category_Neighborhoods = TO_df['Venue_Category'].value_counts()[:6].index.tolist()
top_category_Neighborhoods
```

We will ignore the 'Restaurant' tag as it is very broad. Instead, we will use the 5 most common which are Italian, Japanese, Sushi, Fast Food, and Thai.

```python
top5_Categories = ['Italian Restaurant', 'Japanese Restaurant', 'Sushi Restaurant', 
                       'Fast Food Restaurant', 'Thai Restaurant']
n = 10 # number of neighborhoods
for category in top5_Categories:
    plt.figure(num=None, figsize=(12, 7), dpi = 80, facecolor='w', edgecolor='k')
    plt.title(f'Top {n} Neighborhoods With The Most {category}s')
    get_top_Neighborhoods = venue_count[category].sort_values(ascending=False)[0:n]
    get_top_Neighborhoods = get_top_Neighborhoods.sort_values(ascending=True)
    
    get_top_Neighborhoods.plot.barh(y=category, rot=0)
```

**For each venue category, determine the percentage of entities in each neighborhood**

```python
venue_total = {}
for category in TO_df['Venue_Category']:
    venue_total[category] = venue_count[category].sum()
    
venue_total
```

```python
venue_mean = pd.DataFrame()
for category, total in venue_total.items():
    venue_mean[category] = venue_count[category].apply(lambda x:x / total)

venue_mean.insert(0, 'Neighborhood', 'null')
venue_mean['Neighborhood'] = venue_count['Neighborhood']
venue_mean.head(5)
```

Now, we will create a function and a dataframe that will show us the top 5 venue categories for each neighborhood.

```python
def get_top_venue_Categories(row, num):
    row_categories = row.iloc[1:]
    row_categories_Sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_Sorted.index.values[0:num] #returns sorted df
```

```python
num_of_TVenues = 5   # 5 top venues 

indicators = ['st', 'nd', 'rd'] # nifty little trick

#sort columns by neighborhood top venues 
columns = ['Neighborhood']
for indic in np.arange(num_of_TVenues):
    try: 
        columns.append('{}{} Top Venue Category'.format(indic+1, indicators[indic]))
    except:
        columns.append('{}th Top Venue Category'.format(indic+1))
        
top_venue_categories = pd.DataFrame(columns=columns)
top_venue_categories['Neighborhood'] = venue_mean['Neighborhood']

for ind in np.arange(venue_mean.shape[0]):
    top_venue_categories.iloc[ind, 1:] = get_top_venue_Categories(venue_mean.iloc[ind, :], num_of_TVenues)

top_venue_categories.head(5)
```

Now we will move onto the last section: Clustering, The Visualization of the clusters, and finally examining the contents of clusters

### KMeans Cluster Model for Toronto Neighborhoods

```python
kclusters = 15 # num of clusters 

venue_grouping = venue_mean.drop('Neighborhood', 1)
kmeans = KMeans(n_clusters = kclusters, random_state=0).fit(venue_grouping)
 #display clusters in array 
kmeans.labels_
```

**Now create a new dataframe that mergese the neighborhood location data, top restaurant category by neighborhood, and the label name of the clusters**

```python
top_venue_categories.insert(1,'Cluster Labels', kmeans.labels_)
cluster_Df = geo_data.drop(columns=['Postalcode', 'Borough'])
# merge toronto data with previous dataframe with lat/lng for each neighborhood

cluster_Df = cluster_Df.join(top_venue_categories.set_index('Neighborhood'), on ='Neighborhood').dropna(axis=0)
cluster_Df.reset_index(drop=True, inplace=True)
cluster_Df['Cluster Labels'] = cluster_Df['Cluster Labels'].fillna("0").astype(int)
cluster_Df['Cluster Labels'] = cluster_Df['Cluster Labels'].astype('int64')

cluster_Df.head()
```

### Create A Visualization of The Clusters

```python
folium_map_TO = folium.Map(location=[latitude, longitude], zoom_start=10)

# setup map colors
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

#markers for map
markers_colors = []
for lat, lon, poi, cluster in zip(cluster_Df['Latitude'], cluster_Df['Longitude'], cluster_Df['Neighborhood'],
cluster_Df['Cluster Labels']):
    label = folium.Popup(str(poi) + 'Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker([lat, lon],
                       radius = 5,
                       popup = label,
                       color = rainbow[cluster-1],
                       fill = True,
                       fill_color = rainbow[cluster-1],
                       fill_opacity=0.6).add_to(folium_map_TO)

display(folium_map_TO)
```
